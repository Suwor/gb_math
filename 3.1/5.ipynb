```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(X, Y)
Z2 = X
Z1 = X-10
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)
Z2 = 10*X**2 + 10*Y**2
Z1 = -10*X**2 + -10*Y**2
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
