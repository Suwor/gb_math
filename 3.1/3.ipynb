```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(t), np.sin(t))
plt.grid()
plt.axis([-1,1,-1,1])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(t)*0.5, np.sin(t))
plt.grid()
plt.axis([-1,1,-1,1])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
t = np.linspace(-10, 10, 100)
plt.plot(t, 1/t)
plt.grid()
plt.show()
```
