```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

def pol2cart(rho, phi):
    x = rho * math.cos(round(math.radians(phi), 2))
    y = rho * math.sin(round(math.radians(phi), 2))
    return [x, y]

x, y = pol2cart(5, 45)
plt.axis([0,15,0,10])
plt.grid(True)
plt.plot(x, y, "ro")
```

```python
x = np.linspace(0, 15, 100)
y = x*0 + np.pi/2
plt.polar(y, x)
plt.ylim([0, 16]);
plt.polar(0, 0, "ro")
plt.polar(np.pi/2, 15, "ro")
```
