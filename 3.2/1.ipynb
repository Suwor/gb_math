```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 3*np.pi, 201)
plt.plot(x, np.cos(x - np.pi/2)*0.5+1)
plt.plot(x, np.cos(x - np.pi/3)*2+0.2)
plt.plot(x, np.cos(x - np.pi/4)*5+1.6)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
